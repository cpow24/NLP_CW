```python
#Read first text file
with open(f'gdrive/My Drive/NLP_CW/Data/original/{file_list[0]}') as f:
  doc = f.read()
doc = re.sub('\n', ' ', doc)
```

```python
#tokenize book
tokenizer = BertTokenizerFast.from_pretrained('bert-base-cased')
tokens = get_tokenized_doc(doc, tokenizer)
doc_tokens = flatten(tokens['sentences'])
```

```python
import en_core_web_lg
nlp = en_core_web_lg.load()
nlp.max_length = 5000000
```

```python
#Named entity extraction
ner = nlp(doc)
entity = []
start_idx = []
end_idx = []
ent_type = []

for ent in ner.ents:
  entity.append(ent.text)
  ent_type.append(ent.label_)
  start_idx.append(ent.start_char)
  end_idx.append(ent.end_char)
```

```python
ner_df = pd.DataFrame(list(zip(entity, ent_type, start_idx, end_idx )),
                      columns=['entity', 'entity_type', 'start_idx', 'end_idx'])
```

```python
#Capitalizing entity names
ner_df['entity'] = ner_df['entity'].apply(lambda x: x.capitalize())
```

```python
#Top 15 most frequent characters mentioned
char_list = list(ner_df[ner_df['entity_type'] == 'PERSON']['entity'].value_counts()[:15].index)
```

```python
#Resolve issue of 'PERSON' entities wrongly identified as non-person identities
char_df = ner_df[ner_df['entity'].isin(char_list)]
char_df['entity_type'] = 'PERSON'
```

```python
char_df['entity'].value_counts()
```

```python
#Getting tokenized document as a list of strings
token_doc = utils.inv_map(doc_tokens)
```

```python
'''
Given a list of characters, returns a dictionary of character names and
their corresponding unique IDs.
'''
def get_char_ids(char_list):

  #Dictionary of character IDs
  char_dict = {}

  for id, char in enumerate(char_list):
    char_dict[char] = id + 1
  
  return char_dict
```

```python
#Creating character ID dict
char_dict = get_char_ids(char_list)
```
